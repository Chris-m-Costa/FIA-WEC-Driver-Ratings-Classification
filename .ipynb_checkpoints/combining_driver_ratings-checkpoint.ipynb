```python
import pandas as pd
import numpy as np

from unidecode import unidecode as un

pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 5000)
```

```python
rt_17 = pd.read_csv('wec_entry_lists/driver_ratings_17.csv')
rt_17['year'] = rt_17.loc[:, 'year'] = 2017
```

```python
rt_18 = pd.read_csv('wec_entry_lists/driver_ratings_18.csv')
rt_18['year'] = rt_18.loc[:, 'year'] = 2018
```

```python
rt_19 = pd.read_csv('wec_entry_lists/driver_ratings_19.csv')
rt_19['year'] = rt_19.loc[:, 'year'] = 2019
```

```python
rt_20 = pd.read_csv('wec_entry_lists/driver_ratings_20.csv')
rt_20['year'] = rt_20.loc[:, 'year'] = 2020
```

```python
rt_21 = pd.read_csv('wec_entry_lists/driver_ratings_21.csv')
rt_21['year'] = rt_21.loc[:, 'year'] = 2021
```

```python
rt_22 = pd.read_csv('wec_entry_lists/driver_ratings_22.csv')
rt_22['year'] = rt_22.loc[:, 'year'] = 2022
```

```python
driver_ratings=pd.concat([rt_17, rt_18, rt_19, rt_20, rt_21, rt_22]).drop('Unnamed: 0', axis=1)

# We've added a year column to each of the imported dataframes and joined them together
```

```python
#removing some errant values from the joined dataframe that was breaking the function below. 

driver_ratings = driver_ratings[(driver_ratings.driver_name != 'Test Day Driver #3') & (driver_ratings.driver_name != '"Steve Brooks"')]
driver_ratings = driver_ratings.sort_values('driver_name').reset_index(drop=True)
```

```python
# The function is correcting the capitalization, removing punction (accents and such), 
# and trailing whitespaces from the driver names. The same function will be run against the main dataset to allow easier 
# joining of the driver_ratings info here to the main dataset later.

def fix_driver_names(driver_ratings):
    
    from unidecode import unidecode as un
    
   
    for i in range(len(driver_ratings['driver_name'])):
        try:
            driver_ratings['driver_name'][i] = un(driver_ratings['driver_name'][i])
            driver_ratings['driver_name'][i] = driver_ratings['driver_name'][i].lower()
            driver_ratings['driver_name'][i] = driver_ratings['driver_name'][i].strip()
        except:
            print(i)
        
    return driver_ratings
        
```

```python
driver_ratings = fix_driver_names(driver_ratings)
```
